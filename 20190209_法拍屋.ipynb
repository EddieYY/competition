```python
my_url="http://aomp.judicial.gov.tw/abbs/wkw/WHD2A03.jsp?pageTotal=13&pageSize=15&rowStart=16&saletypeX=1&proptypeX=C52&courtX=TPD&order=odcrm&query_typeX=session&saleno=&hsimun=all&ctmd=all&sec=all&crmyy=&crmid=&crmno=&dpt=&saledate1=&saledate2=&minprice1=&minprice2=&sumprice1=&sumprice2=&area1=&area2=&registeno=&checkyn=all&emptyyn=all&order=odcrm&owner1=&landkd=&rrange=%A4%A3%A4%C0&comm_yn=&stopitem=&courtNoLimit=&pageNow=15&97166F29649F5A2FE8180AD3FF02C722=D074AEC2F3B4A5BDF2F7EF2E25516A5F"
```

```python
total_page = int(my_url.split("pageTotal=")[1].split("&")[0])
```

```python
data = []
for p in range(1, (total_page+1)):
    url_tmp = my_url.split("&pageNow=")[0] + "&pageNow="+ str(p) + '&97166F29649F5A2FE8180AD3FF02C722=D074AEC2F3B4A5BDF2F7EF2E25516A5F'
    uClient =uReq(url_tmp)   #open html and srart to craw 
    page_html = uClient.read()
    uClient.close()
    page_soup= soup(page_html,'html.parser')  # html parser 
    table_ =  page_soup.table.find_all('table')[1]
    rows = table_.find_all('tr')
    s=0  
    if p!=1 :
        s = 1
    for i in range(s, len(rows)):
        row = rows[i] 
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append([ele.replace('\r\n', '').replace('\t', '').replace('\n', '').replace('\u3000', '').replace(' ', '') for ele in cols if ele])
```

```python
df = DataFrame(data)
df.columns = data[0]
df = df[1:]
df
```
